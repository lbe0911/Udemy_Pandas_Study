### 기본 설정되어 있는 행 개수 변경하기

pd.options.display.max_rows - 최대 행 개수 /
pd.options.display.min_rows - 최소 행 개수

데이터 프레임을 불러왔을때 행의 개수가 60개를 넘기면 최소 행 개수인 10개만 보여준다(맨 첫 행부터 아래로 5번째 행, 맨 마지막 행부터 위로 5번째 행)

개수를 변경하고 싶다면 예를 들어 pd.options.display.min_rows=20 이렇게 변경하면 된다.

### 데이터 점검
주요 과제를 시작전 데이터의 상태 점검(결측치 여부, 데이터 정리 등등)

INDEX = 데이터프레임 행의 맨 왼쪽에 위치, 기본은 Rangeindex 0부터 시작~ / 가로 줄
COLUMN = 행과 반대인 열, 예를들어 titanic 데이터프레임의 열은 survived, pcalss, sex, age..등이 있다. / 세로 줄

```python
# info() = 판다스에 가장 중요하다고 볼수 있음. 데이터프레임 요약과 메타 정보를 보여준다.
titanic.info()
# Column : 컬럼명 / Non-Null Count : NULL개수 / Dtype : 데이터유형
```

```python
# describe() = describe메소드는 숫자로된 열에 대한 요약 통계를 반환한다.
titanic.describe()
# count : 개수 / mean : 평균 / std : 표준편차 / min : 최소값 / 25% : 1 사분위 / 50% : 중앙값 / 75% : 3사분위 / max : 최대값
```

```python
# describe() 메소드는 숫자가 아닌 데이터도 분석이 가능하다. 매개변수 include 입력 = "O" 대문자 O
titanic.describe(include="O")
# count : 개수 / unique : 고유값(중복제거) / top : 최빈값 / freq : 빈도수
```
